1. Data Preprocessing Tasks

Before analysis or modeling, ensure data is clean and ready.

Data type correction: Convert order_date, ship_date, and sales (currently object type) to proper formats (datetime and float).

Check for missing values: Identify and handle nulls (though initial check shows none).

Remove duplicates: Verify if duplicate order_id + product_id combinations exist.

Feature engineering:

Extract month, day, and weekday from order_date.

Compute shipping_days = ship_date - order_date.

Calculate profit margin = profit / sales.

Outlier detection: Identify unusual sales, profit, and discount values.

```python
df["order_id"].unique()
```

```python
df["order_date"] = pd.to_datetime(df["order_date"])
df["ship_date"] = pd.to_datetime(df["ship_date"])
```

```python
df["sales"] = df["sales"].str.replace(",","")
```

```python
df["sales"] = df["sales"].astype(float)
```

```python
print(f"datatype of Order_data column {df["order_date"].dtype}")
print(f"Datatype of ship_date column {df["ship_date"].dtype}")
print(f"Datatype of sales column {df["sales"].dtype}")
```

```python
df[["order_id","product_id"]]
```

```python
df["months"] = df["order_date"].dt.month_name()
```

```python
df.drop(columns=["Year"],inplace = True)
```

```python
df.head()
```

```python
df["Day"] = df["order_date"].dt.day_name()
```

```python
df.head()
```

```python
df["weekend"] = df["Day"].apply(lambda ele:1 if ele == "Sunday" else 0)
```

```python
df["weekend"].unique()
```

```python
df["Day"].unique()
```

```python
#counting number of order date wise, month wise, weekend

def count_plot(col):
    plt.figure(figsize=(8,5))
    sns.countplot(data=df,x=df[col],palette="plasma")
    plt.xticks(rotation=45)
    plt.title(f"Number of orders in every {col}")
    plt.tight_layout()
```

```python
for col in ["months","Day","weekend"]:
    count_plot(col)
    plt.show()
```
