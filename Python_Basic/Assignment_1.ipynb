# Data Preprocessing

```python
df["order_id"].unique()
```

```python
df["order_date"] = pd.to_datetime(df["order_date"])
df["ship_date"] = pd.to_datetime(df["ship_date"])
```

```python
#counting number of order date wise, month wise, weekend

def count_plot(col):
    plt.figure(figsize=(8,5))
    sns.countplot(data=df,x=df[col],palette="plasma")
    plt.xticks(rotation=45)
    plt.title(f"Number of orders in every {col}")
    plt.tight_layout()
```

```python
for col in ["months","Day","weekend"]:
    count_plot(col)
    plt.show()
```

# Data Processing / Feature Engineering for ML

```python
# converts categorical columns into numeric for ML model

df_cat = df[['ship_mode', 'segment', 'region', 'category', 'sub_category', 'order_priority']].copy()
```

```python
df_cat.head()
```

```python
for col in df_cat.columns:
    print(f"unique values in {col} columns :",df[col].unique())
    print()
    print("--"*75)
```

```python
def level(ele):
    if ele == "Standard Class":
        return 4
    elif ele == "Second Class":
        return 3
    elif ele == "First Class":
        return 2
    elif ele == "Same Day":
        return 1
    else:
        return 0
```

```python
df_cat["Ship_mode_Level"] = df["ship_mode"].apply(level)
```

```python
df_cat.head()
```

```python
def priority_level(ele):
    if ele == "Critical":
        return 1
    elif ele == "High":
        return 2
    elif ele == "Medium":
        return 3
    elif ele == "Low":
        return 4
```

```python
df_cat["Priority_level"] = df["order_priority"].apply(priority_level)
```

```python
df_cat.head()
```

```python
def seg_lev(ele):
    if ele == "Consumer":
        return 1
    elif ele == "Home Office":
        return 2
    elif ele == "Corporate":
        return 0
```

```python
df_cat["Num_segment"] = df_cat["segment"].apply(seg_lev)
```

```python
df_cat.head()
```

```python
def cat_lev(ele):
    if ele == "Office Supplies":
        return 1.0
    elif ele == "Furniture":
        return 2.0
    elif ele == "Technology":
        return 3.0
```

```python
df_cat["Num_category"] = df["category"].apply(cat_lev)
```

```python
df_cat.head(10)
```

```python
region_num = dict(enumerate(df["region"].unique()))
```

```python
region_num
swap = {val: keys for keys, val in region_num.items()}
swap
```

```python
df_cat["Num_region"] = df_cat["region"].apply(lambda ele:swap[ele])
```

```python
df_cat.head()
```

```python
sub_cat_num = dict(enumerate(df["sub_category"].unique()))
swap_2= {val: key for key,val in sub_cat_num.items()}
```

```python
df_cat["Sub_cat_number"] = df_cat["sub_category"].apply(lambda ele:swap_2[ele])
```

```python
df_cat.head()
```

```python
df_catnum = df_cat[["Ship_mode_Level","Priority_level","Num_segment","Num_category","Num_region","Sub_cat_number"]].copy()
```
